```python
#functionising 
def calc_preds(coeffs, indeps): return (indeps*coeffs).sum(axis=1)
def calc_loss(coeffs, indeps, deps): return torch.abs(calc_preds(coeffs, indeps)-deps).mean()

```

# Gradient Descent Step

```python
coeffs.requires_grad_()
```

# Create Validation and Train

```python
from fastai.data.transforms import RandomSplitter
trn_split,val_split=RandomSplitter(seed=42)(df)
val_split
```

```python
trn_indep,val_indep = t_indep[trn_split],t_indep[val_split]
trn_dep,val_dep = t_dep[trn_split],t_dep[val_split]
len(trn_indep),len(val_indep)
```

```python
#functionising the gradient descent step
def update_coeffs(coeffs, lr):
    coeffs.sub_(coeffs.grad * lr)
    coeffs.grad.zero_()
    
def one_epoch(coeffs, lr):
    loss = calc_loss(coeffs, trn_indep, trn_dep)
    loss.backward()
    with torch.no_grad(): update_coeffs(coeffs, lr)
    print(f"{loss:.3f}", end="; ")
    
def init_coeffs(): return (torch.rand(n_coeff)-0.5).requires_grad_()
```

```python
def train_model(epochs=30, lr=0.01):
    torch.manual_seed(442)
    coeffs = init_coeffs()
    for i in range(epochs): one_epoch(coeffs, lr=lr)
    return coeffs
```

```python
coeffs = train_model(18 , lr=0.1)
```

# Transitioning to neural net

A neural net is essentially just a regression model with more layers of coeffs, and when you get  many layers this becomes deep learning.

Each additional layer of coefficients comes with its own coefficient to multiply that layer by which designates its importance and relationship to target.

I like to think of each layer as representing a characteristic of the relationship between variables and target,(e.g layer one may represent a linear regression model to determine a persons disposition to having some heart disease and layer 2 their current state of health, the combination of the two, as determined by the higher layer of coefficients, will determine the likelyhood that they have said disease) however this kind of interpretation doesnt always hold, infact it often doesnt, and becomes harder with deeper models or more complex issues for obvious reasons. 

This is one argument to not use deep learning unless necessary: linear regression may get similar results and be much easier to interpret as will likely be the case with this data set.

now initialising the coefficients becomes slightly more complex, we need a matrix of size n_coeffs by n_hidden for layer 1 and n_hidden by 1 for layer 2

```python
def init_coeffs(n_hidden=20):
    layer1 = (torch.rand(n_coeff, n_hidden)-0.5)/n_hidden # divide by n_hidden for normalisation purposes
    layer2 = torch.rand(n_hidden, 1)-0.3
    const = torch.rand(1)[0]
    return layer1.requires_grad_(),layer2.requires_grad_(),const.requires_grad_()
```

```python
import torch.nn.functional as F

def calc_preds(coeffs, indeps):
    l1,l2,const = coeffs
    res = F.relu(indeps@l1)
    res = res@l2 + const
    return torch.sigmoid(res)
```

```python
def update_coeffs(coeffs, lr):
    for layer in coeffs:
        layer.sub_(layer.grad * lr)
        layer.grad.zero_()
```
